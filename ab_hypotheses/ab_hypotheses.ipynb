## Приоритизация гипотез

### Приоритизация гипотез способом ICE

```python
#Посчитаем ICE для гипотез и отсортируем их по убыванию приоритета.
hypothesis['ICE'] = (hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']
display(hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False))
```

Наиболее перспективной является гипотеза "Запустить акцию, дающую скидку на товар в день рождения", у нее очень высокие показатели 'confidence' и 'impact' - 9/10, т.е. изменение окажет существенное влияние на пользователей и аналитики очень уверены, что произойдет именно так. Показатель 'efforts' у данной гипотезы 5/10, т.е. тестирование этой гипотезы не слишком дорогое, но и не дешевое.

На второй позиции оказывается гипотеза "Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей", показатель 'impact' 10/10, что означает очень сильно влияние на пользователей, тистирование этой гипотезы окажется чуть дороже предыдущей - 'efforts' 6/10 и аналитики уверенны в ней на 8/10.

Также высоким приоритетом обладает гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", она имеет следующие показатели: 'confidence' 8/10, 'impact' 7/10, 'efforts' 5/10. Все три показателя меньше или равны показателям у двух предыдущих гипотез, поэтому данная гипотеза оказывается только третьей по приоритетности.

### Приоритизация гипотез способом RICE

```python
#Посчитаем RICE для гипотез и отсортируем их по убыванию приоритета.
hypothesis['RICE'] = (hypothesis['reach']*hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']
display(hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False))
```

Самой перспективной гипотезой с двойным отрывом показателя RICE стала гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок", она имеет следующие показатели: 'confidence' 8/10, 'impact' 7/10, 'efforts' 5/10, 'reach' 10/10. Показатель 'reach' оценивает скольких пользователей затронет изменение, которое мы хотим внести, это самый высокий показатель среди всех остальных гипотез.

**Вывод:**
После применения RICE вместо ICE приоритизация гипотез изменилась. На первую позицию вышла гипотеза, которая при приоритезации способом ICE была лишь третьей. Это гипотеза "Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок". Она изначально имела довольно высокие показатели 'confidence', 'impact' и 'efforts', а максимальная оценка 'reach', которая говорит о том, что изменение затронет всех пользователей, позволила данной гипотезе лидировать в фреймворке RICH.

Гипотезы "Запустить акцию, дающую скидку на товар в день рождения" и  "Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей" опустились с первой и второй позиции на пятую и третью соответственно. Это произошло потому, что данные гипотезы не охватывают большое количество людей и получили показатель 'reach' 1 и 3. 

## Анализ A/B-теста

Создадим общую таблицу с кумулятивными данными за каждый день.

```python
#создаем таблицу с датами и группами, которые у нас есть
date_group = orders[['date', 'group']].drop_duplicates().sort_values(by=['date', 'group'])
date_group.head(5)
```

```python
#создаем кумулятивную таблицу по заказам
orders_aggregated = date_group\
.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]\
.agg({'date': 'max','group': 'max','transactionId': 'nunique','visitorId': 'nunique','revenue': 'sum'}),axis=1)\
.sort_values(by=['date', 'group'])
#Получим строки таблицы orders, дата которых меньше или равна дате элемента из date_group, 
#а группа теста равна группе из date_group. Агрегируем значения. Вычислим максимальную дату. 
#Для группы тоже рассчитаем максимум, чтобы столбец 'group' не пропал из итогового вывода. 
#Найдём число уникальных ID заказов и пользователей. Подсчитаем сумму средних чеков.
#Применим методы к каждой строке датафрейма и отсортируем результаты по столбцам.
orders_aggregated.head(5)
```

```python
#создадим аналогичную таблицу по посетителям
visitors_aggregated = date_group\
.apply (lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]\
.agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1)\
.sort_values(by=['date', 'group'])
visitors_aggregated.head(5)
```

```python
#соединяем получившиеся таблицы
cumulative_data = orders_aggregated.merge(visitors_aggregated, on=['date', 'group'])
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
cumulative_data.head(5)
```

```python
#создадим отдельную таблицу для группы А и B
cumulative_data_A = cumulative_data[cumulative_data['group'] == 'A'][['date', 'orders', 'buyers', 'revenue', 'visitors']]
cumulative_data_B = cumulative_data[cumulative_data['group'] == 'B'][['date', 'orders', 'buyers', 'revenue', 'visitors']]
cumulative_data_A.head(5)
```

### График кумулятивной выручки по группам

```python
plt.figure(figsize=(12,6))
plt.plot(cumulative_data_A['date'], cumulative_data_A['revenue'], label='A')
plt.plot(cumulative_data_B['date'], cumulative_data_B['revenue'], label='B')
plt.ylabel('Выручка', fontsize=12)
plt.xlabel('Дата', fontsize=12)
plt.title('График кумулятивной выручки по группам', fontsize=15)
plt.legend(fontsize=12);
```

**Вывод:** выручка по группе А на протяжении всего времени меньше, чем у группы B. При этом между 17 и 21.08.2019 наблюдается резкий скачок в группе B. ### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

Посчитаем статистическую значимость различий в среднем количестве заказов на пользователя между группами по «сырым» данным — без удаления аномальных пользователей.

Соберем данные о всех пользователях (и тех, кто совершал заказ и тех, кто не совершал). 
- 1) Сформируем таблицы (А и B) с уникальными Id пользователя и количеством заказов из таблицы orders.
- 2) Посчитаем разницу между всеми пользователями из таблицы visitors и количеством пользователей с заказами из таблицы orders.
- 3) Соединим получившиеся таблицы друг за другом и получим данные по всем пользователям (в том числе и пользователям без заказа)

```python
orders_by_userA = (
    orders[orders['group'] == 'A'].groupby('visitorId')['transactionId'].nunique().reset_index()
)
orders_by_userA.columns = ['user', 'orders']
orders_by_userB = (
    orders[orders['group'] == 'B'].groupby('visitorId')['transactionId'].nunique().reset_index()
)
orders_by_userB.columns = ['user', 'orders']

no_ordersA = pd.Series(
    0, index=np.arange(visitors[visitors['group']=='A']['visitors'].sum() - len(orders_by_userA['orders'])), name='orders'
)
no_ordersB = pd.Series(
    0, index=np.arange(visitors[visitors['group']=='B']['visitors'].sum() - len(orders_by_userB['orders'])), name='orders'
)

all_ordersA = pd.concat([orders_by_userA['orders'], no_ordersA])
all_ordersB = pd.concat([orders_by_userB['orders'], no_ordersB])
```

```python
#посмотрим получившиеся таблицы
display(all_ordersA.sort_values(ascending=False).head(5))
all_ordersB.sort_values(ascending=False).head(5)
```

Максимальное значение заказов не совпадает с максимальным значением из предыдущей таблицы orders_number, там максимальное значени было 11. Вероятно, это связано с тем, что в данных имеются пользователи, которые попали в обе группы.

Проверим, нормально ли распределены данные с помощью критерия Шапиро-Уилка

```python
#в группе А
alpha = 0.05  # критический уровень статистической значимости

results = st.shapiro(all_ordersA)
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение в группе А: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

```python
#в группе B
results = st.shapiro(all_ordersB)
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение в группе B: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

Распределение не нормально, для проверки статистической значимости различий используем непараметрический тест Манна-Уитни. Сформулируем гипотезы. 
- Нулевая: различий в среднем количестве заказов между группами нет. 
- Альтернативная: различия в среднем количестве заказов между группами есть.

```python
results = st.mannwhitneyu(all_ordersA, all_ordersB) #по умолчанию тест двухсторонний

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: различия в среднем между группами есть')
else:
    print('Не получилось отвергнуть нулевую гипотезу, различий в среднем количестве заказов между группами нет') 

print("Относительный выигрыш в группе B {0:.2f}%".format((all_ordersB.mean() / all_ordersA.mean() - 1)*100))
```

**Вывод:** по не очищенным данным разница между средним количества заказов на пользователя в группе B выше на 13,81%, чем в группе А, разница является статистически значимой.

### Cтатистическая значимость различий в среднем чеке заказа между группами по «сырым» данным

Посчитаем статистическу значимость различий в среднем чеке заказа между группами по 'сырым' данным.

```python
#проверим нормально ли распределены данные в группе А
results = st.shapiro(orders[orders['group']=='A']['revenue'])
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение в группе А: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

```python
#проверим нормально ли распределены данные в группе B
results = st.shapiro(orders[orders['group']=='B']['revenue'])
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение в группе B: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

Распределение не нормально, для проверки статистической значимости различий используем непараметрический тест Манна-Уитни. Сформулируем гипотезы.
- Нулевая гипотеза: различий в среднем чеке между пользователями нет
- Альтернативная гипотеза: различие в среднем чеке меду группами есть

```python
results = st.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue']) #по умолчанию тест двухсторонний

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: различия в среднем между группами есть')
else:
    print('Не получилось отвергнуть нулевую гипотезу, различий в среднем количестве заказов между группами нет') 

print("Относительный выигрыш в группе B {0:.2f}%".format((orders[orders['group']=='B']['revenue'].mean() / orders[orders['group']=='A']['revenue'].mean() - 1)*100))
```

**Вывод:** средний чек в группе B на 25% выше, чем в группе А, однако эта разница не является статистически значимой. Такая большая относительная разница и отсустсвие статистической значимости указывает на наличие выбросов в выборке и их влияение на метрику.

### Статистическая значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным.

Аномальными считаются пользователи, совершившие более 4 заказов на сумму свыше 58233 рублей.

```python
many_orders = orders_number[orders_number['orders'] > limit_orders]['user'] #отберем пользователей более чем с 4 заказами (из двух групп)
expensive_order = orders[orders['revenue'] > limit_revenue]['visitorId'] #отберем пользователей с дорогими заказми
abnormal_users = pd.concat([many_orders, expensive_order]).drop_duplicates() #соберем все в один список
len(abnormal_users) #посмотрим сколько всего пользователей отсеялось
```
