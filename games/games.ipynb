# Проект "Анализ игровой индустрии."

Иинтернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Представлены данные до 2016 года. 

**Цели проекта:** необходимо выявить закономерности, определяющие успешность игры. Сделать ставку на потенциально популярный продукт и спланировать рекламные кампани на 2017 год.

**Задачи проекта:** изучить и подготовить данные, провести исследовательский анализ данных, составить портреть пользователя каждого региона, проверить гипотезы.

**Описание данных:**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB

### Изменение регистра. 

````python
#приведем к нижнему регистру название столбцов
data.columns = data.columns.str.lower()
#также приведем к нижнему регистру содержимое категоральных столбцов 
data['name'] = data['name'].str.lower()
data['platform'] = data['platform'].str.lower()
data['genre'] = data['genre'].str.lower()
data.head()
````

### Обработка пропущенных значений. 

````python
# check
# пропущенные значения бары + значения

def pass_value_barh(df):
    
    fig, ax = plt.subplots(figsize=(19, 6))
    
    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by='space', ascending=True)
            .plot(kind='barh', figsize=(19,6), legend=False, fontsize=16, ax = ax)
            .set_title('Пропущенные значения, %' + "\n", fontsize=22, color='SteelBlue')
        );
#         plt.grid(visible=True, color='grey', linestyle='--', linewidth=0.5, alpha=0.8)
        plt.xlim(0, 100)
    
        for i in ax.patches:
            plt.text(i.get_width()+0.2, i.get_y()+0.1,
                     str(round((i.get_width()), 1)),
                     fontsize=14, 
                     color='black')
    
    except:
        print('пропусков не осталось :) или произошла ошибка в первой части функции')
````

````python
#посмотрим на пропуски в столбце с оценкой критиков, оценкой пользователей и рейтингом ESRB
data.query('critic_score.isnull() & user_score.isnull() & rating.isnull()')
````

Пропуски имеются как для игр, выпущенных давно, так и для новых. Это касается и столбцов с оценкой критиков и пользователей, а также с оценкой рейтинга ESRB. ESRB была основана в 1994 году, а значит игры, выпущенные ранее 1994 года не имеют его. Также рейтинг ESRB не распространяется на Японию. Вероятно, что ситуция с оценкой критиков и пользователей аналогичная - раньше это просто не использовалось. Пропуски в более "свежих" релизах также можно объяснить низкой популярностью игры.

````python
#заполним пропуски значением -1, чтобы они не мешали для дальнейшего анализа
# и не влияли на статистику (-1 потому что некоторые игры имеют рейтинг 0)
data.loc[data['critic_score'].isna(), 'critic_score'] = -1
data.loc[data['user_score'].isna(), 'user_score'] = -1
data.loc[data['rating'].isna(), 'rating'] = 'no_rating'
````

````python
#посмотрим количество пропусков после преобразований
data.isna().sum()
````

````python
#посмотрим уникальные значения в столбце с оценкой пользователей
data['user_score'].unique()
````

В столбце имеется значение 'tbd'. TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. То есть полученное количество оценок у игры мало, поэтому не возмоно проставить рейтинг.

````python
#tbd равноценно пропускам, поэтому его заменим на -1
data.loc[data['user_score']=='tbd', 'user_score'] = -1
````

### Топ-6 платформ за все время.

````python
top_six = data.query('platform == ["ps2", "x360", "ps3", "wii", "ds", "ps", "pc"]')
top = (
    top_six.pivot_table
    (index = 'year_of_release', 
     columns = 'platform', 
     values ='total_sales',
     aggfunc ='sum')
)
top.plot(figsize = (13,7), title='Суммарные продажи по топ-6 платформам')
plt.xlabel('Годы игровых релизов')
plt.ylabel('Продажи')
plt.show()
````

Для построения прогноза на 2017 год возьмем данные за последние три года (т.е. с 2013 по 2016), так как игровая индустрия меняется очень быстро.

### Суммарные продажи по платформам за 2013-2016 г.

При фокусировании на более актуальном периоде времени лидеры поменялись. Однако, чтобы не упустить "новинки" и платформы, актуальные в 2016 году, но не набравшие продаж (так как существуют не долгое время), рассмотрим всех: PS4, PS3, XOne, 3DS, X360, WiiU, PC, PSV, Wii, PSP, DS.

Из графика видно, что в 2016 году по продажам лидируют PS4, XOne и 3DS. Эти три платформы можно рассматривать как потенциально прибыльные. Также в перечень перспективных платформ включим персональные компьютеры (PC), т.к. эта платформа не теряет свою актуальность со временем, хотя и приносит меньше прибыли, чем три другие платформы. ## Проверка гипотез.

Необходимо проверить следующие гипотезы:
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Проверка гипотезы: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

```
H_0: средний пользовательский рейтинг (user_score) платформы XOne = средний пользовательский рейтинг (user_score) платформы PC
H_1: средний пользовательский рейтинг (user_score) платформы XOne ≠ средний пользовательский рейтинг (user_score) платформы PC
alpha = 0.05
```

````python
#применяем гипотезу о равенстве двух средних двух генеральных совокупностей
sample_1 = new_data.loc[(new_data['platform'] == 'xone')&(new_data['user_score']>=0), 'user_score']
sample_2 = new_data.loc[(new_data['platform'] == 'pc')&(new_data['user_score']>=0), 'user_score']

alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print(results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Нулевая гипотеза не отвергается") 
````

Вывод: между средней пользовательской оценкой XOne и PC нет статистически значимой разницы. Вероятность того, что при выборе двух случаных значений из двух выборок они окажутся равны - 14,7%.

Проверка гипотезы: средние пользовательские рейтинги жанров Action и Sports разные;

```
H_0: средний пользовательский рейтинг (user_score) жанра Action = средний пользовательский рейтинг (user_score) жанра Sports
H_1: средний пользовательский рейтинг (user_score) жанра Action ≠ средний пользовательский рейтинг (user_score) жанра Sports
alpha = 0.05
```

````python
#применяем гипотезу о равенстве двух средних двух генеральных совокупностей
sample_3 = new_data.loc[(new_data['genre'] == 'action')&(new_data['user_score']>=0), 'user_score']
sample_4 = new_data.loc[(new_data['genre'] == 'sports')&(new_data['user_score']>=0), 'user_score']

alpha = 0.05

results = st.ttest_ind(sample_3, sample_4)

print(results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
````

Вывод: средние пользовательские рейтинги жанров actions и sports различаются. Вероятность того, что при выборе двух случаных значений из двух выборок они окажутся равны крайне мала.
